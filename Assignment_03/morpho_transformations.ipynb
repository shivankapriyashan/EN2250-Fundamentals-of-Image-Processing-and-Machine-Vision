```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt

img = cv.imread("original.png",0)
assert img is not None



kernel = np.ones((5,5),np.uint8)
erosion =  cv.erode(img,kernel,iterations = 1)
dilation = cv.dilate(erosion,kernel,iterations = 1)

fig, ax = plt.subplots(1,3,figsize=(10,10))
ax[0].imshow(img, cmap='gray')
ax[1].imshow(erosion,cmap='gray')
ax[2].imshow(dilation,cmap='gray')
plt.show()
```

```python
img_opening = cv.imread('opening 2.png',cv.IMREAD_GRAYSCALE)
opening = cv.morphologyEx(img,cv.MORPH_OPEN,kernel)
#Erosion followed by dilation - used to remove noise

erode = cv.erode(img_opening,kernel,iterations=1)
dilate = cv.dilate(erode,kernel,iterations=1)

fig, ax = plt.subplots(1,4,figsize=(10,10))
ax[0].imshow(img_opening,cmap='gray')
ax[1].imshow(erode,cmap='gray')
ax[2].imshow(dilate,cmap='gray')
ax[3].imshow(opening,cmap='gray')
plt.show()
```

```python
img_closing = cv.imread('closing.png',cv.IMREAD_GRAYSCALE)
closing = cv.morphologyEx(img_closing,cv.MORPH_CLOSE,kernel)
# Dilation followed by erosion - used to close small holes inside the foreground object

dilate = cv.dilate(img_closing,kernel,iterations=1)
erode = cv.erode(dilate,kernel,iterations=1)

fig, ax = plt.subplots(1,4,figsize=(10,10))
ax[0].imshow(img_closing,cmap='gray')
ax[1].imshow(dilate,cmap='gray')
ax[2].imshow(erode,cmap='gray')
ax[3].imshow(closing,cmap='gray')
plt.show()
```

```python
gradient = cv.morphologyEx(img,cv.MORPH_GRADIENT,kernel)

fig, ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(img,cmap='gray')
ax[1].imshow(gradient,cmap='gray')
plt.show()
```

```python
tophat = cv.morphologyEx(img,cv.MORPH_TOPHAT,kernel)

fig, ax = plt.subplots(1,3,figsize=(10,10))
ax[0].imshow(img,cmap='gray')
ax[1].imshow(opening,cmap='gray')
ax[2].imshow(tophat,cmap='gray')
plt.show()
```

```python
blackhat = cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernel)

fig, ax = plt.subplots(1,3,figsize=(10,10))
ax[0].imshow(img,cmap='gray')
ax[1].imshow(closing,cmap='gray')
ax[2].imshow(blackhat,cmap='gray')
plt.show()
```

```python
#Using the elliptical structuring element
ellipse = cv.getStructuringElement(cv.MORPH_ELLIPSE,(5,5))
```

```python
img_opening = cv.imread('opening 2.png',cv.IMREAD_GRAYSCALE)
opening = cv.morphologyEx(img,cv.MORPH_OPEN,ellipse)
#Erosion followed by dilation - used to remove noise

erode = cv.erode(img_opening,ellipse,iterations=1)
dilate = cv.dilate(erode,ellipse,iterations=1)

fig, ax = plt.subplots(1,4,figsize=(10,10))
ax[0].imshow(img_opening,cmap='gray')
ax[1].imshow(erode,cmap='gray')
ax[2].imshow(dilate,cmap='gray')
ax[3].imshow(opening,cmap='gray')
plt.show()
```
